```python
# game and shit

import tkinter as tk
from tkinter import simpledialog, messagebox

# how do you do this. how to tell if smth is prime. this does it
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

# gotta get a class in this bih to handle logic
class PrimeGridGame:
    def __init__(self, root):
        self.root = root
        self.grid_size = self.select_difficulty()  # Ask user for grid size
        self.grid = [[None for _ in range(self.grid_size)] for _ in range(self.grid_size)]
        self.current_number = 1
        self.start_square_selected = False
        self.init_ui()
    #3x3 and above. might need to change later if i add scoring idk
    def select_difficulty(self):
        while True:
            try:
                size = int(simpledialog.askstring("Grid Size", "Enter grid size (3 or greater):"))
                if size >= 3:
                    return size
                else:
                    messagebox.showerror("Error", "Grid size must be 3 or greater!")
            except ValueError:
                messagebox.showerror("Error", "Please enter a valid integer!")
    #does this stuff work on every computer?
    def init_ui(self):
        self.root.title("Prime Grid Puzzle")
        self.buttons = [[None for _ in range(self.grid_size)] for _ in range(self.grid_size)]

        # ok finally the grid is the right size
        screen_width = self.root.winfo_screenwidth()
        screen_height = self.root.winfo_screenheight()
        grid_window_size = min(screen_width, screen_height) * 0.6
        button_size = int(grid_window_size / self.grid_size)
        font_size = max(8, int(button_size * 0.4))
        font = ("Helvetica", font_size, "bold")

        # making the buttons square is fucking annoying when you have to put text in them
        for i in range(self.grid_size):
            self.root.grid_rowconfigure(i, minsize=button_size, weight=1)
            self.root.grid_columnconfigure(i, minsize=button_size, weight=1)

        for row in range(self.grid_size):
            for col in range(self.grid_size):
                btn = tk.Button(self.root, text="", font=font, command=lambda r=row, c=col: self.place_number(r, c))
                btn.grid(row=row, column=col, sticky="nsew", padx=1, pady=1)
                self.buttons[row][col] = btn

        # label that goes on bottom of grid
        self.status_label = tk.Label(self.root, text="Select a starting square.", font=("Helvetica", 12))
        self.status_label.grid(row=self.grid_size, column=0, columnspan=self.grid_size)

    def place_number(self, row, col):

        if not self.start_square_selected:
            self.start_game(row, col)
        else:
            # you can only put numbers adjacently
            if self.is_adjacent(row, col) and self.grid[row][col] is None:
                self.grid[row][col] = self.current_number
                self.buttons[row][col].config(text=str(self.current_number), state="disabled")
                self.color_code_square(row, col, self.current_number)
                self.current_number += 1
                self.update_status()

                # check game over. this needs significant work if someone softlocks. i havent figured out how to do that yet but like. i could.
                if self.current_number > self.grid_size ** 2:
                    self.status_label.config(text="Game over!")

    def start_game(self, row, col):
        self.grid[row][col] = self.current_number
        self.buttons[row][col].config(text=str(self.current_number), state="disabled")
        self.color_code_square(row, col, self.current_number)
        self.start_square_selected = True
        self.current_number += 1
        self.status_label.config(text="Place numbers in adjacent squares.")

    def is_adjacent(self, row, col):
        # oh right i actually have to write the check if its adjacent damn it
        for r in range(-1, 2):
            for c in range(-1, 2):
                if 0 <= row + r < self.grid_size and 0 <= col + c < self.grid_size:
                    if self.grid[row + r][col + c] == self.current_number - 1:
                        return True
        return False

    def color_code_square(self, row, col, number):
        if is_prime(number):
            self.buttons[row][col].config(bg="light blue")
        else:
            self.buttons[row][col].config(bg="light coral")

    def update_status(self):
        self.status_label.config(text=f"Current number: {self.current_number}")

# YEAH BABEY
if __name__ == "__main__":
    root = tk.Tk()
    game = PrimeGridGame(root)
    root.mainloop()
```
